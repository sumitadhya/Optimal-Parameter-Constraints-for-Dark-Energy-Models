# Flat XCDM

```python
def H_theory(z, H0, Obh2, Och2,wx):
    """Calculates the theoretical Hubble parameter H(z) for a flat LambdaCDM model."""
    Ovh2 = 0.06 / 93.14
    h = H0 / 100.0
    Om0 = (Obh2 + Och2 + Ovh2) / h**2
    Ez = np.sqrt(Om0 * (1 + z)**3 + (1 - Om0) * (1 + z)**(3 * (1 + wx)))
    return H0 * Ez
```

```python
c = 3e5  # km/s

def E_flat(z, H0, Obh2, Och2, wx):
    Ovh2 = 0.06 / 93.14
    h = H0 / 100.0
    Om0 = (Obh2 + Och2 + Ovh2) / h**2
    return np.sqrt(Om0 * (1 + z)**3 + (1 - Om0) * (1 + z)**(3 * (1 + wx)))

def Dh_flat(z, H0, Obh2, Och2, wx):
    return c / (H0 * E_flat(z, H0, Obh2, Och2, wx))

def Dc_flat(z, H0, Obh2, Och2, wx):
    result, _ = scipy.integrate.quad(lambda zp: 1.0/E_flat(zp, H0, Obh2, Och2, wx), 0, z)
    return (c / H0) * result

def Dm_flat(z, H0, Obh2, Och2, wx):
    # Flat LambdaCDM: Transverse comoving = line-of-sight comoving
    return Dc_flat(z, H0, Obh2, Och2, wx)

def Da_flat(z, H0, Obh2, Och2, wx):
    return Dm_flat(z, H0, Obh2, Och2, wx) / (1 + z)

def Dv_flat(z, H0, Obh2, Och2, wx):
    # Volume-averaged distance for BAO
    return (Dm_flat(z, H0, Obh2, Och2, wx)**2 * z * Dh_flat(z, H0, Obh2, Och2, wx))**(1/3)

def rs_flat(H0, Obh2, Och2, wx):
    Ovh2 = 0.06 / 93.14
    p1 = np.exp(-72.3 * (Ovh2 + 0.0006)**2)
    p2 = (Obh2)**0.12807 * (Obh2 + Och2)**0.25351
    return 55.154 * p1 / p2

```

```python
def log_likelihood1_old(theta, z, H_obs, H_err):
    H0, Obh2, Och2, wx = theta
    H_th = H_theory(z, H0, Obh2, Och2, wx)
    res = H_obs - H_th
    chi21 = np.sum((res / H_err)**2)
    log_norm = np.sum(np.log(2.0 * np.pi * (H_err**2)))

    return -0.5 * (chi21 + log_norm)
```

```python
def log_likelihood1_new(theta, z, H_obs, cov_total):
    H0, Obh2, Och2, wx = theta
    H_th = H_theory(z, H0, Obh2, Och2, wx)
    res = H_obs - H_th
    inv_cov = np.linalg.inv(cov_total)
    chi2 = np.dot(res, np.dot(inv_cov, res))
    # Normalization (multivariate normal)
    log_norm = np.linalg.slogdet(2.0 * np.pi * cov_total)[1]
    return -0.5 * (chi2 + log_norm)
```

```python
def log_likelihood2_old(theta, z_bao, bao_val, cov_matrix):
    H0, Obh2, Och2, wx = theta
    
    N = len(z_bao)
    D_th = np.zeros(N)
    rs_fid = 147.5

    rs_values = rs_flat(H0, Obh2, Och2, wx) * np.ones(N)  # flat model, constant rs for all z

    # Dv, Da, Dm, Dh functions use (z, H0, Obh2, Och2, wx)
    D_th[0] = Dv_flat(z_bao[0], H0, Obh2, Och2, wx) * rs_fid / rs_values[0]      # Dv/rs @ z=0.122
    D_th[7] = Da_flat(z_bao[7], H0, Obh2, Och2, wx) / rs_values[7]               # Da/rs @ z=0.81
    for i in [1,3,5,8,10]:       # Dm/rs indices
        D_th[i] = Dm_flat(z_bao[i], H0, Obh2, Och2, wx) / rs_values[i]
    for i in [2,4,6,9,11]:       # Dh/rs indices
        D_th[i] = Dh_flat(z_bao[i], H0, Obh2, Och2, wx) / rs_values[i]

    res = bao_val - D_th

    inv_cov = np.linalg.inv(cov_matrix)
    chi2 = np.dot(res, np.dot(inv_cov, res))
    log_norm = np.linalg.slogdet(2.0 * np.pi * cov_matrix)[1]

    return -0.5 * (chi2 + log_norm)
```

```python
def log_likelihood2_new(theta, z_bao, bao_val, cov_matrix):
    H0, Obh2, Och2, wx = theta
    
    N = len(z_bao)
    D_th = np.zeros(N)
    rs_fid = 147.5

    rs_values = rs_flat(H0, Obh2, Och2, wx) * np.ones(N)  # flat model, constant rs for all z

    # Dv, Da, Dm, Dh functions use (z, H0, Obh2, Och2, wx)
    D_th[0] = Dv_flat(z_bao[0], H0, Obh2, Och2, wx) * rs_fid / rs_values[0]      # Dv/rs @ z=0.122
    #D_th[7] = Da_flat(z_bao[7], H0, Obh2, Och2, wx) / rs_values[7]               # Da/rs @ z=0.81
    for i in [1,3,5,7,8,10]:       # Dm/rs indices
        D_th[i] = Dm_flat(z_bao[i], H0, Obh2, Och2, wx) / rs_values[i]
    for i in [2,4,6,9,11]:       # Dh/rs indices
        D_th[i] = Dh_flat(z_bao[i], H0, Obh2, Och2, wx) / rs_values[i]

    res = bao_val - D_th

    inv_cov = np.linalg.inv(cov_matrix)
    chi2 = np.dot(res, np.dot(inv_cov, res))
    log_norm = np.linalg.slogdet(2.0 * np.pi * cov_matrix)[1]

    return -0.5 * (chi2 + log_norm)
```

```python
def log_prior(theta):
    H0, Obh2, Och2, wx = theta
    if 0.0 < Obh2 < 1.0 and 0.0 < Och2 < 1.0 and -5.0 < wx < 0.33:
        return 0.0
    return -np.inf
```

```python
def log_probability_old(theta, z, H_obs, H_err, z_bao, bao_val, cov_matrix):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood1_old(theta, z, H_obs, H_err)+log_likelihood2_old(theta, z_bao, bao_val, cov_matrix)
```

```python
def log_probability_new(theta, z, H_obs, cov_total,z_bao, bao_val, cov_matrix):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood1_new(theta, z, H_obs, cov_total)+log_likelihood2_new(theta, z_bao, bao_val, cov_matrix)
```

```python
def negative_log_probability_old(theta, z, H_obs, H_err,z_bao, bao_val, cov_matrix):
    logp = log_prior(theta)
    if not np.isfinite(logp):
        return 1e10
    return -log_likelihood1_old(theta, z, H_obs, H_err) - log_likelihood2_old(theta, z_bao, bao_val, cov_matrix)

theta_init = [70.0, 0.022, 0.12, -1.0]

result = minimize(
    negative_log_probability_old,
    theta_init,
    args=(z_data_old, Hz_data_old, sigma_data_old, z_oldbao, oldbao_val, cov_matrix_old),
    method='Nelder-Mead'
)

best_fit_params_old = result.x
print("Best-fit parameters (rounded): H0 = {:.3f}, Obh2 = {:.3f}, Och2 = {:.3f}, wx = {:.3f}".format(*best_fit_params_old))
print("Minimized negative log likelihood:", result.fun)
```

```python
def negative_log_probability_new(theta, z, H_obs, cov_total,z_bao, bao_val, cov_matrix):
    logp = log_prior(theta)
    if not np.isfinite(logp):
        return 1e10
    return -log_likelihood1_new(theta, z, H_obs, cov_total) - log_likelihood2_new(theta, z_bao, bao_val, cov_matrix)

theta_init = [70.0, 0.022, 0.12, -1.0]

result = minimize(
    negative_log_probability_new,
    theta_init,
    args=(z_data_new, Hz_data_new, cov_total, z_newbao, newbao_val, cov_matrix_new),
    method='Nelder-Mead'
)

best_fit_params_new = result.x
print("Best-fit parameters (rounded): H0 = {:.3f}, Obh2 = {:.3f}, Och2 = {:.3f}, wx = {:.3f}".format(*best_fit_params_new))
print("Minimized negative log likelihood:", result.fun)
```

```python
def H_theory(z, H0, Obh2, Och2,wx):
    """Calculates the theoretical Hubble parameter H(z) for a flat LambdaCDM model."""
    Ovh2 = 0.06 / 93.14
    h = H0 / 100.0
    Om0 = (Obh2 + Och2 + Ovh2) / h**2
    Ez = np.sqrt(Om0 * (1 + z)**3 + (1 - Om0) * (1 + z)**(3 * (1 + wx)))
    return H0 * Ez
```

```python
c = 3e5  # km/s

def E_flat(z, H0, Obh2, Och2, wx):
    Ovh2 = 0.06 / 93.14
    h = H0 / 100.0
    Om0 = (Obh2 + Och2 + Ovh2) / h**2
    return np.sqrt(Om0 * (1 + z)**3 + (1 - Om0) * (1 + z)**(3 * (1 + wx)))

def Dh_flat(z, H0, Obh2, Och2, wx):
    return c / (H0 * E_flat(z, H0, Obh2, Och2, wx))

def Dc_flat(z, H0, Obh2, Och2, wx):
    result, _ = scipy.integrate.quad(lambda zp: 1.0/E_flat(zp, H0, Obh2, Och2, wx), 0, z)
    return (c / H0) * result

def Dm_flat(z, H0, Obh2, Och2, wx):
    # Flat LambdaCDM: Transverse comoving = line-of-sight comoving
    return Dc_flat(z, H0, Obh2, Och2, wx)

def Da_flat(z, H0, Obh2, Och2, wx):
    return Dm_flat(z, H0, Obh2, Och2, wx) / (1 + z)

def Dv_flat(z, H0, Obh2, Och2, wx):
    # Volume-averaged distance for BAO
    return (Dm_flat(z, H0, Obh2, Och2, wx)**2 * z * Dh_flat(z, H0, Obh2, Och2, wx))**(1/3)

def rs_flat(H0, Obh2, Och2, wx):
    Ovh2 = 0.06 / 93.14
    p1 = np.exp(-72.3 * (Ovh2 + 0.0006)**2)
    p2 = (Obh2)**0.12807 * (Obh2 + Och2)**0.25351
    return 55.154 * p1 / p2

```

```python
def log_likelihood1_old(theta, z, H_obs, H_err):
    H0, Obh2, Och2, wx = theta
    H_th = H_theory(z, H0, Obh2, Och2, wx)
    res = H_obs - H_th
    chi21 = np.sum((res / H_err)**2)
    log_norm = np.sum(np.log(2.0 * np.pi * (H_err**2)))

    return -0.5 * (chi21 + log_norm)
```

```python
def log_likelihood1_new(theta, z, H_obs, cov_total):
    H0, Obh2, Och2, wx = theta
    H_th = H_theory(z, H0, Obh2, Och2, wx)
    res = H_obs - H_th
    inv_cov = np.linalg.inv(cov_total)
    chi2 = np.dot(res, np.dot(inv_cov, res))
    # Normalization (multivariate normal)
    log_norm = np.linalg.slogdet(2.0 * np.pi * cov_total)[1]
    return -0.5 * (chi2 + log_norm)
```

```python
def log_likelihood2_old(theta, z_bao, bao_val, cov_matrix):
    H0, Obh2, Och2, wx = theta
    
    N = len(z_bao)
    D_th = np.zeros(N)
    rs_fid = 147.5

    rs_values = rs_flat(H0, Obh2, Och2, wx) * np.ones(N)  # flat model, constant rs for all z

    # Dv, Da, Dm, Dh functions use (z, H0, Obh2, Och2, wx)
    D_th[0] = Dv_flat(z_bao[0], H0, Obh2, Och2, wx) * rs_fid / rs_values[0]      # Dv/rs @ z=0.122
    D_th[7] = Da_flat(z_bao[7], H0, Obh2, Och2, wx) / rs_values[7]               # Da/rs @ z=0.81
    for i in [1,3,5,8,10]:       # Dm/rs indices
        D_th[i] = Dm_flat(z_bao[i], H0, Obh2, Och2, wx) / rs_values[i]
    for i in [2,4,6,9,11]:       # Dh/rs indices
        D_th[i] = Dh_flat(z_bao[i], H0, Obh2, Och2, wx) / rs_values[i]

    res = bao_val - D_th

    inv_cov = np.linalg.inv(cov_matrix)
    chi2 = np.dot(res, np.dot(inv_cov, res))
    log_norm = np.linalg.slogdet(2.0 * np.pi * cov_matrix)[1]

    return -0.5 * (chi2 + log_norm)
```

```python
def H_theory(z, H0, Obh2, Och2, Ok, wx):
    return H0 * E_nonflat(z, H0, Obh2, Och2, Ok, wx)
```

```python
def E_nonflat(z, H0, Obh2, Och2, Ok, wx):
    Ovh2 = 0.06 / 93.14  # massive neutrino contribution kept fixed
    h = H0 / 100.0
    Om0 = (Obh2 + Och2 + Ovh2) / h**2
    return np.sqrt(Om0 * (1 + z)**3 + Ok * (1 + z)**2 + (1 - Om0 - Ok) * (1 + z)**(3 * (1 + wx)))
```

```python
# Comoving horizon scale Dh(z) = c/H(z)
def Dh_nonflat(z, H0, Obh2, Och2, Ok, wx):
    return c / (H0 * E_nonflat(z, H0, Obh2, Och2, Ok, wx))

# Line-of-sight comoving distance
def Dc_nonflat(z, H0, Obh2, Och2, Ok, wx):
    result, _ = scipy.integrate.quad(lambda zp: 1.0 / E_nonflat(zp, H0, Obh2, Och2, Ok, wx), 0, z)
    return (c / H0) * result

# Transverse comoving distance (includes curvature cases)
def Dm_nonflat(z, H0, Obh2, Och2, Ok, wx):
    dc = Dc_nonflat(z, H0, Obh2, Och2, Ok, wx)
    dh = c / H0
    if abs(Ok) < 1e-8:
        return dc
    elif Ok > 0:
        sqrtk = np.sqrt(Ok)
        return dh / sqrtk * np.sinh(sqrtk * dc / dh)
    else:
        sqrtk = np.sqrt(-Ok)
        return dh / sqrtk * np.sin(sqrtk * dc / dh)

# Angular diameter distance
def Da_nonflat(z, H0, Obh2, Och2, Ok, wx):
    return Dm_nonflat(z, H0, Obh2, Och2, Ok, wx) / (1 + z)

# Volume-averaged distance
def Dv_nonflat(z, H0, Obh2, Och2, Ok, wx):
    return (Dm_nonflat(z, H0, Obh2, Och2, Ok, wx)**2 * z * Dh_nonflat(z, H0, Obh2, Och2, Ok, wx)) ** (1 / 3)

# Sound horizon proxy (same fitting form as earlier cell)
def rs_nonflat(H0, Obh2, Och2, Ok, wx):
    Ovh2 = 0.06 / 93.14
    p1 = np.exp(-72.3 * (Ovh2 + 0.0006)**2)
    p2 = (Obh2)**0.12807 * (Obh2 + Och2)**0.25351
    return 55.154 * p1 / p2
```
